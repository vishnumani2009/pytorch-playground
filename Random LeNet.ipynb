## LeNet on Random data

```python
#import basic requirements
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
import torch.optim as optim
import torch.utils.data as utils
import numpy as np
```

```python
#Basic network definition
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(1, 10, kernel_size=5)
        self.conv2 = nn.Conv2d(10, 20, kernel_size=5)
        self.conv2_drop = nn.Dropout2d()
        self.fc1 = nn.Linear(320, 120)
        self.fc1_drop = nn.Dropout()
        self.fc2 = nn.Linear(120, 10)
        
    def forward(self, x):
        x = F.relu(F.max_pool2d(self.conv1(x), 2))
        x = F.relu(F.max_pool2d(self.conv2_drop(self.conv2(x)), 2))
        x = x.view(-1, self.num_flat_features(x))
        x = F.relu(self.fc1_drop(self.fc1(x)))
        x = F.relu(self.fc2(x))
        return F.log_softmax(x)
    
    def num_flat_features(self, x):
        size = x.size()[1:] # all dimensions except the batch dimension
        num_features = 1
        for s in size:
            num_features *= s
        return num_features
```

```python
#network defintiion
net=Net()
print(net)
```

```python
#network paramteres
params=list(net.parameters())
#print params[0]

X_np=[np.random.rand(2,32,32) for i in range(100)]
Y_np=[np.random.rand(1,10) for i in range(100)]
#print Y_np
X=torch.stack([torch.Tensor(i) for i in X_np])
Y=torch.stack([torch.Tensor(i) for i in Y_np])
```

```python
optimizer=optim.SGD(net.parameters(),lr=0.01)
criterion=torch.nn.MSELoss()
```
