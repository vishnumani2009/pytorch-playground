### Numpy Neural network

#### This a numpy implementation of three layer neural network of (1000 input ,100 hidden ,10 ouput) configuration

```python
import numpy as np

#layer size specification of the three layer neural network
batchsize=64
inputsize=1000
hiddensize=100
outputsize=10
lr=0.00000006
#create input and output
X=np.random.randn(batchsize,inputsize)
Y=np.random.randn(batchsize,outputsize)

#print(X.shape,Y.shape)
W1=np.random.randn(inputsize,hiddensize)
W2=np.random.randn(hiddensize,outputsize)
#print(W1.shape,W2.shape)

for i in range(3500):
    
    #input to hidden layer (h_i)
    h_i=X.dot(W1)
    h_relu=np.maximum(h_i,0)
    
    #input to last(output) layer before loss calculation (l_i)
    ypred=h_relu.dot(W2)
    
    #loss layer computation
    loss=np.square(ypred-Y).sum()
    
    print loss,i
    #gradient computations
    grad_y_pred = 2.0 * (ypred - Y)
    grad_w2 = h_relu.T.dot(grad_y_pred)
    grad_h_relu = grad_y_pred.dot(W2.T)
    grad_h = grad_h_relu.copy()
    grad_h[h_i < 0] = 0
    grad_w1 = X.T.dot(grad_h)
    #print grad_w1,grad_w2
    
    W1 -= lr * grad_w1
    W2 -= lr * grad_w2                     
                         
```

#### This a cuda tensor implementation of three layer neural network of (1000 input ,100 hidden ,10 ouput) configuration

```python
import numpy as np
import torch

dtype= torch.cuda.FloatTensor
#layer size specification of the three layer neural network
batchsize=64
inputsize=1000
hiddensize=100
outputsize=10
lr=0.00000006
#create input and output
X=torch.randn(batchsize,inputsize)
Y=torch.randn(batchsize,outputsize)

#print(X.shape,Y.shape)
W1=torch.randn(inputsize,hiddensize)
W2=torch.randn(hiddensize,outputsize)
#print(W1.shape,W2.shape)

for i in range(3500):
    
    #input to hidden layer (h_i)
    h_i=X.mm(W1)
    h_relu=h_i.clamp(min=0)
    
    #input to last(output) layer before loss calculation (l_i)
    ypred=h_relu.mm(W2)
    
    #loss layer computation
    loss = (ypred - Y).pow(2).sum()
    
    print loss,i
    #gradient computations
    grad_y_pred = 2.0 * (ypred - Y)
    grad_w2 = h_relu.t().mm(grad_y_pred)
    grad_h_relu = grad_y_pred.mm(W2.t())
    grad_h = grad_h_relu.clone()
    grad_h[h_i < 0] = 0
    grad_w1 = X.t().mm(grad_h)
    #print grad_w1,grad_w2
    
    W1 -= lr * grad_w1
    W2 -= lr * grad_w2                     
                         
```
